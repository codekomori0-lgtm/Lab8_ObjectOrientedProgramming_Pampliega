```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def __str__(self):
        status = "Available" if self.is_available else "Borrowed"
        return f"{self.title:20} | {self.author:15} | {status}"

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))

    def list_books(self):
        print(f"\n{'Title':20} | {'Author':15} | {'Status'}")
        print("-" * 50)
        for book in self.books:
            print(book)

    def borrow(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                if book.is_available:
                    book.is_available = False
                    print(f"✔️ You have successfully borrowed '{book.title}'.")
                    return
                else:
                    print(f"❌ '{book.title}' is already checked out.")
                    return
        print("❌ Book not found.")

    def return_b(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                book.is_available = True
                print(f"✔️ Thank you for returning '{book.title}'.")
                return
        print("❌ This book does not belong to our library.")

# --- Jupyter Testing Cell ---
lib = Library()
lib.add_book("Digital Design", "Morris Mano")
lib.add_book("Electromagnetics", "Sadiku")
lib.list_books()
lib.borrow("Digital Design")
lib.list_books()
```

```python
class Employee:
    def __init__(self, name, rate, hours):
        self.name = name
        self.rate = max(0, rate)
        self.hours = max(0, hours)

    def calculate_payroll(self):
        # Logic: 160 hours base, 1.5x for anything over
        if self.hours > 160:
            base_pay = 160 * self.rate
            overtime_hrs = self.hours - 160
            overtime_pay = overtime_hrs * (self.rate * 1.5)
        else:
            base_pay = self.hours * self.rate
            overtime_pay = 0
            
        total = base_pay + overtime_pay
        
        print(f"\nPayroll for: {self.name}")
        print(f"Base Salary: P{base_pay:,.2f}")
        print(f"Overtime:    P{overtime_pay:,.2f}")
        print(f"Total Gross: P{total:,.2f}")

# --- Jupyter Testing Cell ---
emp1 = Employee("Ram Anthony", 500, 170) # 10 hours overtime
emp1.calculate_payroll()
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

class ShoppingCart:
    def __init__(self):
        self.items = []
        self.total_price = 0

    def add_to_cart(self, product, quantity):
        if quantity <= product.stock:
            self.items.append({"name": product.name, "price": product.price, "qty": quantity})
            product.stock -= quantity
            self.total_price += product.price * quantity
            print(f"✔️ Added {quantity}x {product.name} to cart.")
        else:
            print(f"❌ Insufficient stock for {product.name}!")

    def show_cart(self):
        print("\n--- Your Shopping Cart ---")
        for item in self.items:
            print(f"{item['qty']}x {item['name']} @ ${item['price']} each")
        print(f"Total Amount Due: ${self.total_price:,.2f}")

# --- Jupyter Testing Cell ---
# Predefined Inventory
laptop = Product("MacBook", 1500, 5)
mouse = Product("Magic Mouse", 80, 10)

cart = ShoppingCart()
cart.add_to_cart(laptop, 1)
cart.add_to_cart(mouse, 2)
cart.show_cart()
```
