```python
# Step 1: Define a class with private attributes 
class BankAccount: 
    def __init__(self, account_holder, balance): 
        self.account_holder = account_holder 
        self.__balance = balance  # Private attribute 

    def deposit(self, amount): 
        self.__balance += amount 
        return f"New balance: ${self.__balance}" 

    def get_balance(self): 
        return f"Balance: ${self.__balance}" 
# Step 2: Create an object and test encapsulation 
account = BankAccount("Alice", 5000) 
# Step 3: Deposit money and display the updated balance 
print(account.deposit(1500)) 
print(account.get_balance()) 
```

```python
# Step 1: Define the StudentRecord class
class StudentRecord:
    def __init__(self, student_name):
        self.student_name = student_name
        self.__grades = []  # Private attribute (a list to store grades)

    # Step 2: Method to add a grade
    def add_grade(self, grade):
        if 0 <= grade <= 100:  # Simple validation
            self.__grades.append(grade)
            return f"Grade {grade} added for {self.student_name}."
        else:
            return "Invalid grade. Please enter a value between 0 and 100."

    # Step 3: Method to retrieve all grades
    def get_all_grades(self):
        return f"{self.student_name}'s Grades: {self.__grades}"

# --- Testing Encapsulation ---

# Create an object
student = StudentRecord("Ram Anthony E. Pampliega")

# Add several grades using the method
student.add_grade(85)
student.add_grade(92)
student.add_grade(88)

# Display the output
print(student.get_all_grades())

# Proof of Encapsulation:
# If you try to print(student.__grades), Python will raise an AttributeError.
```

```python
# Step 1: Define a base class 
class Person: 
    def __init__(self, name): 
        self.name = name 
    
    # This must be indented to belong to the class!
    def introduce(self): 
        return f"Hi, I'm {self.name}." 

# Step 2: Define a subclass that inherits from Person 
class Student(Person): 
    def introduce(self): 
        return f"Hi, I'm {self.name}, and I'm a student." 

# Step 3: Create objects and call methods 
person = Person("Ram") 
student = Student("Pampliega") 

print(person.introduce()) 
print(student.introduce())
```

```python
# Step 1: Define the base class
class Device:
    def __init__(self, device_name):
        self.device_name = device_name

    def power_on(self):
        return f"The {self.device_name} is now powered ON."

# Step 2: Define the subclass that overrides the method
class Smartphone(Device):
    # We don't need a new __init__ if we aren't adding new attributes, 
    # it automatically uses the one from Device.
    
    def power_on(self):
        return f"{self.device_name} is booting up... loading OS... Welcome!"

# Step 3: Instantiate objects
generic_gadget = Device("Oscilloscope")
my_phone = Smartphone("iPhone 15")

# Step 4: Display outputs
print(generic_gadget.power_on())
print(my_phone.power_on())
```

```python
# Step 1: Define a base class 
class Shape: 
    def area(self): 
        return "Calculating area..." 

# Step 2: Define subclasses with their implementations 
class Circle(Shape): 
    def __init__(self, radius): 
        self.radius = radius 
    
    def area(self): 
        # Using Pi * r^2
        return 3.14 * (self.radius ** 2) 

class Rectangle(Shape): 
    def __init__(self, length, width): 
        self.length = length 
        self.width = width 
    
    def area(self): 
        # Using Length * Width
        return self.length * self.width 

# Step 3: Create objects and call their methods 
circle = Circle(5) 
rectangle = Rectangle(4, 6) 

print(f"Circle area: {circle.area()}")      # Output: 78.5
print(f"Rectangle area: {rectangle.area()}")  # Output: 24
```

```python
# Step 1: Define the base class
class Device:
    def power_on(self):
        return "Generic device is starting up..."

# Step 2: Define subclasses with unique startup messages
class Smartphone(Device):
    def power_on(self):
        return "Smartphone: Screen glowing... Vibrating... Home screen ready."

class Laptop(Device):
    def power_on(self):
        return "Laptop: Fan spinning... BIOS check complete... Windows/Linux desktop loaded."

# Step 3: Define the polymorphic function
def start_device(device):
    """
    This function accepts any object that has a power_on() method.
    This is the core of Polymorphism!
    """
    print(device.power_on())

# Step 4: Create instances and test
my_phone = Smartphone()
my_laptop = Laptop()

# Passing different objects to the SAME function
start_device(my_phone)
start_device(my_laptop)
```

```python
# Step 1: Define the class with a static method
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        # The formula: F = (C * 9/5) + 32
        return (celsius * 9/5) + 32

# Step 2: Get user input
# We use float() since temperatures are often not whole numbers
user_celsius = float(input("Enter temperature in Celsius: "))

# Step 3: Call the static method directly using the class name
# Note: We don't need to create an object like 'conv = TemperatureConverter()'
fahrenheit = TemperatureConverter.celsius_to_fahrenheit(user_celsius)

# Step 4: Print the result
print(f"{user_celsius}°C is equal to {fahrenheit}°F")
```

```python
# --- PART 1: Temperature Converter ---
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

# Getting user input for temperature
c_input = float(input("Enter temperature in Celsius: "))
f_result = TemperatureConverter.celsius_to_fahrenheit(c_input)
print(f"{c_input}°C is equal to {f_result}°F")

print("-" * 30) # Separator line

# --- PART 2: Vector with Operator Overloading ---
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Overloading the - operator (Using the correct __sub__ method)
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    # Making the output readable
    def __str__(self):
        return f"({self.x}, {self.y})"

# Creating Vector instances
v1 = Vector(10, 20)
v2 = Vector(5, 8)

# Demonstrating polymorphism via operators
v_sum = v1 + v2
v_diff = v1 - v2

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Addition (v1 + v2): {v_sum}")
print(f"Subtraction (v1 - v2): {v_diff}")
```

```python
from abc import ABC, abstractmethod

# --- Part 1: Temperature Converter (Static Method) ---
class TemperatureConverter:
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

# User interaction
c_val = float(input("Enter temperature in Celsius: "))
print(f"Result: {TemperatureConverter.celsius_to_fahrenheit(c_val)}°F")

print("-" * 30)

# --- Part 2: Abstraction (Appliance Hierarchy) ---
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        """This is a requirement for any subclass."""
        pass

class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

class Refrigerator(Appliance):
    def turn_on(self):
        return "Refrigerator is now cooling."

# Step 3: Instantiate and Test
wm = WashingMachine()
ref = Refrigerator()

print(wm.turn_on())
print(ref.turn_on())
```

```python
 # Step 1: Define multiple classes 
class A: 
    def show(self): 
        return "Class A" 
class B(A): 
        def show(self): 
         return "Class B" 
class C(A): 
    def show(self): 
        return "Class C" 
class D(B, C): 
    def show(self):
        return "Class D"  # Overrides the inherited show() method
# Step 2: Create an object of class D and call the method 
obj = D() 
print(obj.show())  # Output depends on MRO 
```

```python
# Step 1: Define the base class
class Base:
    def show(self):
        return "Message from Base"

# Step 2: Define intermediate classes
class A(Base):
    def show(self):
        return "Message from Class A"

class B(Base):
    def show(self):
        return "Message from Class B"

class C(Base):
    def show(self):
        return "Message from Class C"

# Step 3: Define class D inheriting from A, B, and C
# The order inside the parentheses (A, B, C) matters!
class D(A, B, C):
    pass

# Step 4: Test and Explore
obj = D()
print(f"Output of obj.show(): {obj.show()}")

print("\n--- Method Resolution Order (MRO) ---")
# Using mro() to see the search path
for cls in D.mro():
    print(cls)
```
